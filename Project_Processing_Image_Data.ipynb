Resizing the image using Pillow Library

```python
from PIL import Image
```

```python
img = Image.open('/content/Car.jpg')
img_resized = img.resize((200,200))
```

```python
img_resized.save('Car_resized.jpg')
```

```python
img_res = mpimg.imread('/content/Car_resized.jpg')
img_res_plot = plt.imshow(img_res)
plt.show()
```

```python
print(img_res.shape)
```

Converting RGB images to Grayscale images using OpenCV

```python
# Importing OpenCV Library

import cv2
```

```python
img = cv2.imread('/content/Car.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
Grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(Grayscale_image)
```

```python
Grayscale_image.shape
```

cv2.imshow() will display the image but it will not work in Google Colab
So,
from google.colab.patches import cv2_imshow

```python
from google.colab.patches import cv2_imshow
```

```python
# Displaying the image in Grayscale

cv2_imshow(Grayscale_image)
```

```python
# Saving the Grayscale image

cv2.imwrite('Car_Grayscale.jpg',Grayscale_image)
```
